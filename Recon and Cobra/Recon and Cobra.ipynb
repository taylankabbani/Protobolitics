### Imprting Json file

```python
with urllib.request.urlopen("https://raw.githubusercontent.com/sehir-bioinformatics-database-lab/metabolitics/master/metabolitics/datasets/network_models/recon2.json") as url:
    data = json.loads(url.read().decode())
```

### Exploring Recon

```python
print("Recon items:")
for i in data:
    print(i, end = ', ')
```
