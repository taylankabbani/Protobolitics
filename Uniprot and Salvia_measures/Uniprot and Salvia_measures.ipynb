## Protein Measures of samples salvia

```python
# Importing The data
salvia_proteomics = pd.read_excel("SalivaProteomics.XLSX", header = 0).loc[:,'G1.1.Itraq1.LIBRA.113.ratio':'gene'].set_index('gene').reset_index()
```

```python
# Remove missing values from the dataset
salvia_proteomics.dropna(inplace=True)
salvia_proteomics.reset_index(drop=True, inplace=True)
salvia_proteomics.head()
```

```python
salvia_proteomics.info()
```

```python
# Retrurning gene equivalance in HGNC dict:
def hgnc_finder(gene_symbol):
    try:
        return uniprot_HGNC[gene_symbol]
    except Exception:
        return False
```

```python
# This function creats a dict for a sample with hgnc id as key and protein measurements as value
def hgnc_measures(sample_name):
    hgnc_dict = {}
    gene_list = list(salvia_proteomics["gene"])
    for i in range(len(gene_list)):
        hgnc_id = hgnc_finder(gene_list[i])
        # Only including gene symbols that have match in uniprot_HGNC
        if hgnc_id:
            hgnc_dict[hgnc_id] = float(salvia_proteomics[[sample_name]].loc[i])
    return hgnc_dict
```

```python
#list of hgnc measures for all patients
Protein_measures = [hgnc_measures(sample) for sample in salvia_proteomics.drop('gene', axis = 1).columns]
```
